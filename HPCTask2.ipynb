```python
%%writefile FileHandler.c

#include <stdio.h>
#include <stdlib.h>
#include <string.h>

#define BUFFER_SIZE 256

struct MatrixSize {
    int x;
    int y;
};

struct Matrix {
    struct MatrixSize size;
    double* values;
};

struct MatrixSize* charArrayIsMatrixSize (char * characters, int size) {
    struct MatrixSize* mtxSize = (struct MatrixSize*) malloc( sizeof(struct MatrixSize) );
    int splitCount = 0;
    char temp[BUFFER_SIZE];
    strcpy(temp, characters);

    char* split = strtok(temp, ",");
    while (split != NULL) {
        double parsed = atof(split);
        if ( parsed > 0 )
        {
            if (splitCount == 0) {
                mtxSize->x = parsed;
            } else if (splitCount == 1) {
                mtxSize->y = parsed;
            } else {
                return NULL;
            }
            splitCount++;
        }

        split = strtok(NULL, ",");
    }

    return mtxSize;
}

int charArrayIsBlank (char * characters, int size) {
    return characters[0] == '\n';
}

double* parseLineOfDoubles(char* lineCharacters, struct MatrixSize mtxSize) {
    double* lineValues = malloc (sizeof(double) * mtxSize.y);
    int lineYCount = 0;

    const char splitter[2] = ",";
    char* split = strtok(lineCharacters, splitter);
    while (split) {
        double parsed = atof(split);
        if ( parsed > 0 ) {
            lineValues[lineYCount] = parsed;
            lineYCount += 1;
        }
        split = strtok(NULL, splitter);
    }

    return lineValues;
}

int getFileMatrixCount(char* fileName) {
    FILE * filePointer;
    filePointer = fopen(fileName, "r");
    if (filePointer == NULL) {
        printf("Unable to load file '%s'\n", fileName);
        return 0;
    }

    int matrixCount = 0;
    char readBuffer[BUFFER_SIZE];
    while ( fgets(readBuffer, BUFFER_SIZE, filePointer) != NULL ) {
        if ( charArrayIsBlank(readBuffer, BUFFER_SIZE) > 0 ) {
            matrixCount++;
        }
    }
    matrixCount++;

    fclose(filePointer);

    return matrixCount;
}

struct Matrix** loadFromFile (char * fileName) {
    if (!fileName) {
        printf("Unable to find file '%s'\n", fileName);
        return NULL;
    }

    int totalMatrixCount = getFileMatrixCount(fileName);

    FILE * filePointer;
    filePointer = fopen(fileName, "r");
    if (filePointer == NULL) {
        printf("Unable to load file '%s'\n\n", fileName);
        return NULL;
    }

    struct Matrix** matrices = malloc( sizeof(struct Matrix) * totalMatrixCount );

    struct MatrixSize* currentMatrixSize = (struct MatrixSize*) malloc(sizeof(struct MatrixSize));
    currentMatrixSize->x = currentMatrixSize-> y = 0;

    double* currentMatrixValues = NULL;
    int currentMatrixCount = 0;
    int currentMatrixCurrentLineCount = 0;

    char readBuffer[BUFFER_SIZE];
    while ( fgets(readBuffer, BUFFER_SIZE, filePointer) != NULL ) {
        struct MatrixSize* mtxSize = charArrayIsMatrixSize(readBuffer, BUFFER_SIZE);
        if ( mtxSize != NULL && currentMatrixSize->x == 0 && currentMatrixSize->y == 0 ) {
            currentMatrixSize = mtxSize;

        } else if ( charArrayIsBlank(readBuffer, BUFFER_SIZE) > 0 ) {

            struct Matrix* thisMatrix = malloc( sizeof(struct Matrix) );
            thisMatrix->values = currentMatrixValues;
            thisMatrix->size = *currentMatrixSize;

            matrices[currentMatrixCount] = thisMatrix;
            currentMatrixCount++;
            currentMatrixCurrentLineCount = 0;
            currentMatrixSize->x = currentMatrixSize->y = 0;
            currentMatrixValues = NULL;
        } else {
            if (currentMatrixValues == NULL) {
                currentMatrixValues = malloc( sizeof(double) * currentMatrixSize->y);
            }
            double* values = parseLineOfDoubles(readBuffer, *currentMatrixSize);
            for (int i = 0; i < currentMatrixSize->y; i++) {
                int index = (currentMatrixCurrentLineCount * currentMatrixSize->y) + i;
                currentMatrixValues[index] = values[i];
            }
            currentMatrixCurrentLineCount++;
        }
    }

    struct Matrix* thisMatrix = malloc( sizeof(struct Matrix) );
    thisMatrix->values = currentMatrixValues;
    thisMatrix->size = *currentMatrixSize;

    matrices[currentMatrixCount] = thisMatrix;

    currentMatrixCount++;

    fclose(filePointer);

    return matrices;
}

int saveMatricesToFile ( char* filePath, struct Matrix** matrices, int matricesCount) {
    FILE* saveToFile = fopen(filePath, "w");
    if (saveToFile == NULL)
    {
        printf("Unable to open filePath '%s'. Check it isn't open\n\n", filePath);
        return 0;
    }

    for( int i = 0; i < matricesCount; i++ ) {
        struct Matrix matrix = *(matrices[i]);

        fprintf(saveToFile, "%d,%d\n", matrix.size.x, matrix.size.y);

        for (int j = 0; j < matrix.size.x; j++) {
            int rowIndex = (j * matrix.size.y);
            for( int k = 0; k < matrix.size.y; k++ ) {
                fprintf(saveToFile, "%f", matrix.values[rowIndex + k]);
                if (k < matrix.size.y - 1) {
                    fprintf(saveToFile, ",");
                }
            }
            fprintf(saveToFile, "\n");
        }
        fprintf(saveToFile, "\n");
    }

    fclose(saveToFile);
    return 1;
}
```

```python
%%writefile matrixMultipication.c

#include <stdlib.h>
#include <stdio.h>
#include <pthread.h>
#include "FileHandler.c"

struct Matrix** Matrices;

struct ThreadArgs {
    int threadIndex;
    int dotProductCount;
    struct Matrix* resultMatrixInfo;
    double* matrixAValues;
    double* matrixBValues;
};

void printMatrix(double* matrix, int xCount, int yCount);
void* calculate(void* param);
int canMultiplyMatrices(struct Matrix a, struct Matrix b);

void* calculate(void* param) {
    struct ThreadArgs* args = (struct ThreadArgs*)param;
    int row = args->threadIndex / args->resultMatrixInfo->size.y;
    int col = args->threadIndex % args->resultMatrixInfo->size.y;

    if (row < args->resultMatrixInfo->size.x && col < args->resultMatrixInfo->size.y) {
        double total = 0;
        for (int i = 0; i < args->dotProductCount; i++) {
            double aVal = args->matrixAValues[row * args->dotProductCount + i];
            double bVal = args->matrixBValues[i * args->resultMatrixInfo->size.y + col];
            total += aVal * bVal;
        }
        args->resultMatrixInfo->values[row * args->resultMatrixInfo->size.y + col] = total;
    }
    return NULL;
}

int main(int argc, char* argv[]) {
    char* filePath = "./matrix_data.txt";
    if (argc > 1) {
        filePath = argv[1];
    }

    char* outputFilePath = "./matrix_result.txt";
    if (argc > 2) {
        outputFilePath = argv[2];
    }

    int totalMatrixCount = getFileMatrixCount(filePath);
    Matrices = loadFromFile(filePath);

    printf("Successfully loaded '%d' matrices from file '%s'", totalMatrixCount, filePath);
    struct Matrix** allFinalMatrices = malloc(sizeof(struct Matrix*) * totalMatrixCount);
    int successfulMatrixCount = 0;

    for (int i = 0; i < totalMatrixCount; i += 2) {
        struct Matrix matrixA = *(Matrices[i]);
        struct Matrix matrixB = *(Matrices[i + 1]);

        if (canMultiplyMatrices(matrixA, matrixB)) {
            struct Matrix* resultMatrix = malloc(sizeof(struct Matrix));
            resultMatrix->size.x = matrixA.size.x;
            resultMatrix->size.y = matrixB.size.y;
            resultMatrix->values = malloc(sizeof(double) * resultMatrix->size.x * resultMatrix->size.y);

            int threadCount = resultMatrix->size.x * resultMatrix->size.y;
            pthread_t* threadIds = malloc(sizeof(pthread_t) * threadCount);
            struct ThreadArgs* args = malloc(sizeof(struct ThreadArgs) * threadCount);

            for (int j = 0; j < threadCount; j++) {
                args[j].threadIndex = j;
                args[j].dotProductCount = matrixA.size.y;
                args[j].resultMatrixInfo = resultMatrix;
                args[j].matrixAValues = matrixA.values;
                args[j].matrixBValues = matrixB.values;
                pthread_create(&threadIds[j], NULL, calculate, &args[j]);
            }

            for (int j = 0; j < threadCount; j++) {
                pthread_join(threadIds[j], NULL);
            }

            printf("Result of Matrix[%d] * Matrix[%d] =\n\n", i, i + 1);
            printMatrix(resultMatrix->values, resultMatrix->size.x, resultMatrix->size.y);

            allFinalMatrices[successfulMatrixCount++] = resultMatrix;

            free(threadIds);
            free(args);
        } else {
            printf("Unable to multiply Matrices[%d] '%dx%d' and Matrices[%d] '%dx%d'\n\n", i, matrixA.size.x, matrixA.size.y, i + 1, matrixB.size.x, matrixB.size.y);
        }
    }

    int success = saveMatricesToFile(outputFilePath, allFinalMatrices, successfulMatrixCount);
    if (success == 1) {
        printf("Successfully saved '%d' matrices into file '%s'\n\n", successfulMatrixCount, outputFilePath);
    } else {
        printf("Unable to save to file '%s'\n\n", outputFilePath);
    }

    // Free all matrices
    for (int i = 0; i < successfulMatrixCount; i++) {
        free(allFinalMatrices[i]->values);
        free(allFinalMatrices[i]);
    }
    free(allFinalMatrices);

    return 0;
}

int canMultiplyMatrices(struct Matrix a, struct Matrix b) {
    return a.size.y == b.size.x;
}

void printMatrix(double* matrix, int xCount, int yCount) {
    for (int i = 0; i < xCount; i++) {
        for (int j = 0; j < yCount; j++) {
            printf("%f ", matrix[i * yCount + j]);
        }
        printf("\n");
    }
    printf("\n");
}
```
